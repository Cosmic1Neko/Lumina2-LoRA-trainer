```python
# 1. 加载 metadata.json  
in_json = "/root/autodl-tmp/metadata-final2.json"
with open(in_json, "r", encoding="utf-8") as f:  
    metadata = json.load(f)  
      
# 2. 创建 BaseSubset 实例（模拟训练配置）  
subset = BaseSubset(  
    image_dir="/root/autodl-tmp/metadata-final2.json",  
    alpha_mask=False,  
    num_repeats=1,  
    shuffle_caption=True,            
    caption_separator=",",          
    keep_tokens=0,                  
    keep_tokens_separator="|||",    
    secondary_separator=None,  
    enable_wildcard=True,           
    color_aug=False,  
    flip_aug=False,  
    face_crop_aug_range=None,  
    random_crop=False,  
    caption_dropout_rate=0.1,       
    caption_dropout_every_n_epochs=0,  
    caption_tag_dropout_rate=0.0,  
    caption_prefix=None, 
    caption_suffix=None,            
    token_warmup_min=1,             
    token_warmup_step=0,            
    custom_attributes=None,  
    validation_seed=None,  
    validation_split=0.0,  
    resize_interpolation=None,  
)  
      
 # 3. 创建 BaseDataset 实例  
dataset = BaseDataset(  
    resolution=(1280, 1280), 
    network_multiplier=1.0,  
    debug_dataset=True,  
    resize_interpolation='lanczos', 
) 

# 设置必要的参数  
dataset.current_epoch = 1  
dataset.current_step = 100  
dataset.max_train_steps = 1000  
```

```python
import torch
from safetensors.torch import save_file

# 1. 设置文件路径
input_path = "/root/autodl-tmp/model/neta-lumina-alpha-aes-full-a5n-ep2-s58714.pth"  # 你的 .pth 文件路径
output_path = "/root/autodl-tmp/model/neta-lumina-alpha-aes-full-a5n-ep2-s58714.safetensors" # 输出文件名

# 2. 加载 .pth 文件
# map_location="cpu" 可以防止在没有 GPU 的机器上报错，也能节省显存
print(f"正在加载 {input_path} ...")
payload = torch.load(input_path, map_location="cpu")

# 3. 检查数据结构
# 有些 pth 文件直接就是权重字典，有些则包含 'state_dict', 'epoch' 等键
if "state_dict" in payload:
    print("检测到 state_dict 键，正在提取权重...")
    payload = payload["state_dict"]

# 4. 保存为 safetensors
print(f"正在保存为 {output_path} ...")
save_file(payload, output_path)
print("转换完成！")
```
